```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv("DAAEHDataset.csv")
df
```

```python
model = SentimentIntensityAnalyzer()
```

```python
def extract_score(text):
	score = model.polarity_scores(text)
	compound = score['compound']
	sentiment = 'neutral'
	
	if compound >= 0.05:
		sentiment = "positive"
	elif compound <= -0.05:
		sentiment = "negative"

	return sentiment
```

```python
df["sentiment"] = df['Feedback'].apply(extract_score)
df
```

```python
df.to_csv('DAAEHDataset.csv', index=False)
```
