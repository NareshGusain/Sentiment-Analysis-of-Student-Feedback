```python
import pandas as pd
```

```python
df = pd.read_csv('Sentiment Analysis on Student Feedback.csv')
df
```

```python
df['Faculty '].unique()
```

```python
def FacultyintToName(n):
    if n == 1:
        return "Prof. Sunil"
    elif n == 2:
        return "Prof. Reha"
    elif n == 3:
        return "Prof. Ujwal"
    else:
        return "Prof. Kanishik"
```

```python
df['Faculty '] = df['Faculty '].map(FacultyintToName)
```

```python
df = df.drop('Timestamp', axis=1)
```

```python
df
```

```python
df['Course Code'].unique()
```

```python
def Gender(st:str):
    if st == "Prof. Reha":
        return "Female"
    else:
        return "Male"
    
```

```python
df['Gender'] = df["Faculty "].apply(Gender)
df
```

```python
df['Study Hours (per week)'].unique()
```

```python
import re

def extract_numbers(text: str):
    # Find all sequences of digits in the string
    numbers = re.findall(r'\d+', text)
    # Return the first number found, or 0 if no numbers are found
    return int(numbers[0]) if numbers else 0

```

```python
df["Study Hours (per week)"] = df['Study Hours (per week)'].apply(extract_numbers)
df
```

```python
df = df.drop('Unnamed: 11', axis=1)
```
